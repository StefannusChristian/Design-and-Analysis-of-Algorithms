Buatlah suatu algoritma untuk mendeteksi apakah dua buah string *saling anagram* atau tidak.

Suatu string ```s1``` merupakan sebuah anagram dari string ```s2``` jika setiap karakter di ```s1``` dapat disusun ulang urutannya sehingga menjadi ```s2```.

Contoh: string ```study``` adalah anagram dari string ```dusty```, namun string ```tolong``` bukan merupakan anagram dari string ```lontong```.

Ketentuan kompleksitas waktu: $\mathcal{O}(n\log n)$,

Ketentuan kompleksitas ruang: $\mathcal{O}(n)$,

di mana $n$ merupakan ukuran string terpanjang dari antara dua string yang sedang diuji.


````python
class Quicksort:
    def quick_sort(self, start, end, arr):
        if start < end:
            partition_pos = self.partition(start, end, arr)
            self.quick_sort(start, partition_pos, arr)
            self.quick_sort(partition_pos + 1, end, arr)

    def partition(self, start, end, arr):
        pivot = arr[start]
        i = start + 1
        j = end - 1
        print(pivot,i,j)

        while i < j:
            while arr[i] <= pivot and i < end - 1:
                i += 1

            while arr[j] >= pivot and j > start:
                j -= 1

            if i < j:
                arr[i], arr[j] = arr[j], arr[i]

        if arr[j] < pivot:
            arr[j], arr[start] = arr[start], arr[j]

        return j


class Anagram:
    def is_anagram(self, s1, s2, sorter=Quicksort()):
        # di bikin lowercase semua karena python case sensitive
        for i in s1:
            i = i.lower()
        for i in s2:
            i = i.lower()
        # kalau hasil dari string 1 yang sudah di-sort = string 2 yang sudah di-sort
        # maka dia pasti anagram karena kalau hasil sort nya sama pasti string ke dua
        # bisa dibentuk dari huruf" yang ada di string pertama
        s1 = [letter for letter in s1]
        s2 = [letter for letter in s2]
        pjg_s1 = len(s1)
        pjg_s2 = len(s2)
        sorter.quick_sort(0, pjg_s1, s1)
        sorter.quick_sort(0, pjg_s2, s2)
        print(s1,'ahahhahaha')
        if s1 == s2:
            return True
        return False


class AnagramPrinter:
    def print_anagram(self, a=Anagram(), sorter=Quicksort()):
        s1, s2 = 'study', 'dusty'
        s3, s4 = 'tolong', 'lontong'
        s1 = [letter for letter in s1]
        s2 = [letter for letter in s2]
        s3 = [letter for letter in s3]
        s4 = [letter for letter in s4]
        pjg_s1, pjg_s2, pjg_s3, pjg_s4 = len(s1), len(s2), len(s3), len(s4)
        print('is_anagram (study,dusty)')
        if a.is_anagram(s1, s2):
            print(f'sorted {s1} = sorted {s2}')
            (sorter.quick_sort(0, pjg_s1, s1))
            (sorter.quick_sort(0, pjg_s2, s2))
            print(f'{(s1)} = {(s2)}')
            s1, s2 = 'study', 'dusty'
            print(f'{s1} and {s2} is anagram!\n')
        else:
            print(f'sorted {(s1)} != sorted {(s2)}')
            (sorter.quick_sort(0, pjg_s1, s1))
            (sorter.quick_sort(0, pjg_s2, s2))
            print(f'{s1} != {s2}')
            s1, s2 = 'study', 'dusty'
            print(f'{s1} and {s2} is not an anagram!')

        print('is_anagram (tolong,lontong)')
        if a.is_anagram(s3, s4):
            print(f'sorted {(s3)} = sorted {(s4)}')
            (sorter.quick_sort(0, pjg_s3, s3))
            (sorter.quick_sort(0, pjg_s4, s4))
            print(f'{s3} = {s4}')
            s3, s4 = 'tolong', 'lontong'
            print(f'{s3} and {s4} is anagram!')
        else:
            print(f'sorted {(s3)} != sorted {(s4)}')
            (sorter.quick_sort(0, pjg_s3, s3))
            (sorter.quick_sort(0, pjg_s4, s4))
            print(f'{s3} != {s4}')
            s3, s4 = 'tolong', 'lontong'
            print(f'{s3} and {s4} is not an anagram!')


if __name__ == '__main__':
    ap = AnagramPrinter()
    ap.print_anagram()
````

Ketentuan kompleksitas ruang: $\mathcal{O}(n)$.

````python
class Heap:
    # Min Heap --> Child lebih besar dari parent
    # makin kebawah makin besar
    def bottom_up_parent_dominance(self, i, l):
        # Tambahin None di depan list biar sifat-sifat representasi array
        # dari sebuah heap seperti yang diajarkan di-kelas
        if l[0] is not None:
            l.insert(0, None)
        child_index = i
        parent_index = i//2
        # Base case kalau i itu root atau = 1
        # atau ketika parent nya sudah lebih kecil dari child
        if i == 1 or l[parent_index] < l[child_index]:
            l.remove(None)
            return
        # Prosedur Bottom Up
        # Swap kalau parent lebih besar dari child kemudian
        # rekursi parent index nya lakukan terus sampai i itu root
        if l[parent_index] > l[child_index]:
            l[parent_index], l[child_index] = l[child_index], l[parent_index]
            self.bottom_up_parent_dominance(parent_index, l)

    def top_down_parent_dominance(self, i, l):
        # Tambahin None di depan list biar sifat-sifat representasi array
        # dari sebuah heap seperti yang diajarkan di-kelas
        if l[0] is not None:
            l.insert(0, None)
        # simpan index parent left dan right
        parent = i
        left = 2*i
        right = (2*i)+1
        # Base case kalau left = daun
        if left >= len(l):
            l.remove(None)
            return
        # Base case kalau right = daun
        # kalau right = daun masih harus cek sama anak kiri nya dulu
        # kalau dia lebih kecil dari parent maka di swab sama parent
        if right >= len(l):
            if l[parent] > l[left]:
                l[left], l[parent] = l[parent], l[left]
            l.remove(None)
            return
        # Base case kalau parent sudah lebih kecil dari kedua anaknya
        if l[parent] < l[left] and l[parent] < l[right]:
            l.remove(None)
            return
        # Prosedur Top Down
        # Kalau parent lebih kecil dari anaknya yang terkecil
        # maka cek apakah right < left, kalau iya maka kita tukar
        # anak kanan dengan parent kemudian panggil lagi fungsi
        # top down parent dominance dengan i = right
        if l[parent] > min(l[left], l[right]):
            if l[right] < l[left]:
                l[right], l[parent] = l[parent], l[right]
                self.top_down_parent_dominance(right, l)
            # kalau left lebih kecil dari parent maka tukar left dengan parent
            # kemudian top down parent dominance dengan i - left
            else:
                l[left], l[parent] = l[parent], l[left]
                self.top_down_parent_dominance(left, l)

    def heapify(self, l):
        # Tambahin None di depan list biar sifat-sifat representasi array
        # dari sebuah heap seperti yang diajarkan di-kelas
        if l[0] is not None:
            l.insert(0, None)
        n = len(l)
        for i in range(n//2, 0, -1):
            self.top_down_parent_dominance(i, l)

    def insert_key(self, v, l):
        # Tambahin None di depan list biar sifat-sifat representasi array
        # dari sebuah heap seperti yang diajarkan di-kelas
        if l[0] is not None:
            l.insert(0, None)
        l.append(v)
        self.bottom_up_parent_dominance(len(l)-1, l)

    def delete_index(self, i, l):
        # Tambahin None di depan list biar sifat-sifat representasi array
        # dari sebuah heap seperti yang diajarkan di-kelas
        if l[0] is not None:
            l.insert(0, None)
        # Simpan elemen terakhir
        last = l[-1]
        # Swab elemen terakhir dengan index yang ingin di delete
        l[i], last = last, l[i]
        # Buang elemen terakhir --> ga akan merusak struktur heap
        l.pop()
        # Top Down Parent Dominance dari index yang ingin di delete
        self.top_down_parent_dominance(i, l)


class HeapPrinter:
    def print_bottom_up_parent_dominance(self, h=Heap()):
        l = [None, 16, 20, 8, 11, 15, 3, 19, 5, 10, 2]
        print('Bottom Up Parent Dominance')
        print(f'Array = {l}')
        n = 5
        h.bottom_up_parent_dominance(n, l)
        print(f'After Bottom Up {n} = {l}\n')

    def print_top_down_parent_dominance(self, h=Heap()):
        l = [None, 11, 16, 2, 14, 15, 9, 1, 17, 3, 12]
        print('Top Down Parent Dominance')
        print(f'Array = {l}')
        n = 5
        h.top_down_parent_dominance(n, l)
        print(f'After Top Down {n} = {l}\n')

    def print_heapify(self, h=Heap()):
        print('Heapify!')
        l = [None, 16, 20, 8, 11, 15, 3, 19, 5, 10, 2]
        print(f'Normal List = {l}')
        h.heapify(l)
        print(f'Heapify List = {l}\n')

    def print_insert_key(self, h=Heap()):
        v = 100
        l = [None, 16, 20, 8, 11, 15, 3, 19, 5, 10, 2]
        print(f'Insert {v}')
        print(f'Heap List = {l + [v]}')
        h.insert_key(v, l)
        print(f'After {v} is inserted = {l}\n')

    def print_delete_index(self, h=Heap()):
        i = 8
        l = [None, 16, 20, 8, 11, 15, 3, 19, 5, 10, 2]
        print(f'Delete Index {i}')
        print(f'Heap List = {l}')
        h.delete_index(i, l)
        print(f'After Index {i} is deleted = {l}')


if __name__ == '__main__':
    hp = HeapPrinter()
    hp.print_bottom_up_parent_dominance()
    hp.print_top_down_parent_dominance()
    hp.print_heapify()
    hp.print_insert_key()
    hp.print_delete_index()
````

````python
class Horspoll:
    # Tugas 3.1
    # hanya mendaftarkan huruf dari patern
    def toshifttable(self, pattern, shifttable={}):
        n = len(pattern)
        # telusuri dari belakang tapi tidak ambil huruf terakhir
        for i in range(2, n+1):
            # pattern[n-i] itu supaya dia telusuri dari belakang
            if pattern[n-i] not in shifttable:
                shifttable[pattern[n-i]] = i-1
        return shifttable

    # Tugas 3.2
    def horspoll(self, kata, pattern):
        # buat shifttable
        kata = kata.upper()
        pattern = pattern.upper()
        shifttable = self.toshifttable(pattern)
        shift = 0
        print(shifttable)
        panjang_pattern = len(pattern)
        panjang_kata = len(kata)
        # lompat itu penanda lompatan yang akan digeser berdasarkan c shifttable
        lompat = panjang_pattern
        print('Pemeriksaan awal:')
        while lompat < panjang_kata+1:
            if shift != 0:
                print(f'Shift #{shift}')
            print(kata)
            print(' '*(lompat - panjang_pattern) + pattern)
            shift += 1
            found = True
            c = kata[lompat-1]
            # telusuri dari belakang
            for i in range(panjang_pattern-1, -1, -1):
                # kalo pas digeser sudah lewat batas
                if lompat - panjang_pattern + i >= panjang_kata:
                    found = False
                    break
                # kalo ada mismatch
                if pattern[i] != kata[lompat - panjang_pattern + i]:
                    found = False
                    # geser sesuai shifttable
                    if c in shifttable:
                        lompat += shifttable[c]
                    # kalo nda ada dishift table artinya geser semuanya
                    else:
                        lompat += panjang_pattern
                    break
            # kalo pas telusuri sampai akhir variabel foundnya nda berubah artinya stringnya sudah match
            if found:
                print(f'{pattern} ketemu pada index ke {lompat-panjang_pattern}.')
                # geser sesuai shifttable
                if c in shifttable:
                    lompat += shifttable[c]
                # kalo nda ada dishift table artinya geser semuanya
                else:
                    lompat += panjang_pattern
            print()
        if not found:
            print(f'-1\n')


class HorspoolPrinter:
    def print_contoh_diatas(self, sm=Horspoll()):
        print('Contoh diatas')
        return sm.horspoll('SHDOKOSPJANHHAUVIISDJEANCHAUVINPOSLBSJEANCHAUVINSKFL', 'jeanchauvin')
    # Tugas 3.3

    def print_contoh_worstcase(self, sm=Horspoll()):
        print('Contoh worstcase')
        return sm.horspoll('bbbbbbbbbbbbbbbbbbbbbb', 'CBB')


if __name__ == '__main__':
    hp = HorspoolPrinter()
    hp.print_contoh_diatas()
    hp.print_contoh_worstcase()
````

Dalam tugas ini, Anda akan mengimplementasikan Algoritma Rabin-Karp untuk menyelesaikan masalah String Matching dalam konteks bioinformatika. 

Sebuah genome sequence merupakan sebuah string yang setiap karakternya dapat diambil dari himpunan karakter $\{\mathtt{A}, \mathtt{G}, \mathtt{T}, \mathtt{C}\}$ yang masing-masing mewakili 4 jenis basa yang menyusun molekul DNA.

Diberikan sebuah genome sequence yang disimpan oleh variabel string ```genome``` sepanjang $n$ dan sebuah pola basa nukleotida yang disimpan oleh variabel ```pattern``` sepanjang $m$. Anda diminta untuk mencari **semua** lokasi kemunculan ```pattern``` pada ```genome``` dengan menggunakan Algoritma Rabin-Karp.

Untuk setiap posisi pengecekan substring dari ```genome```, Anda perlu mengoutputkan setiap hash-key value dari substring tersebut.

Contoh:
```
genome = AGTCGTTAGAT
pattern = CGTTA
```
Kurang lebih diharapkan algoritma dapat menghasilkan output sebagai demikian.
```
value("CGTTA") = 872

Iterasi 1. substring: AGTCG, value =  109

Iterasi 2. substring: GTCGT, value =	438

Iterasi 3. substring: TCGTT, value =	730

Iterasi 4. substring: CGTTA, value =	872
Pattern is found!

Iterasi 5. substring: GTTAG, value =	417

Iterasi 6. substring: TTAGA, value =	644

Iterasi 7. substring: TAGAT, value =	530
```

*Catatan: Output di atas hanya salah satu metode dari berbagai cara mendesain fungsi hashingnya. Anda tidak harus mengikuti fungsi hashing seperti di atas.*

Ketentuan kompleksitas waktu: $\mathcal{O}(mn)$, dengan average case $\Theta(n)$.

Ketentuan kompleksitas ruang: $\mathcal{O}(m+n)$.

````python
class RabinKarp:
    def myhash(self, string, kamus):
        val = 0
        for i in range(len(string)):
            val += kamus[string[-i-1]]*(4**i)
        return val

    def find(self, genome, pattern, kamus):
        m = len(pattern)
        findval = self.myhash(pattern, kamus)
        print(f'value("{pattern}") = {findval}')

        iter = 1
        cur = self.myhash(genome[:m], kamus)
        print(f'\nIterasi {iter}. substring: {genome[:m]}, value = {cur}')

        for i in range(1, len(genome) - m+1):
            iter += 1
            cur = (cur - kamus[genome[i-1]]*(4**(m-1))) * 4 + kamus[genome[i+m-1]]
            print(f'\nIterasi {iter}. substring: {genome[iter-1:m+iter-1]}, value = {cur}')

            # Ga Langsung Break Karena Pattern Nya Bisa > 1
            if cur == findval:
                print('Pattern is found!')
 

if __name__ == '__main__':
    rb = RabinKarp()
    kamus = {'A': 0,
             'G': 1,
             'T': 2,
             'C': 3}
    genome = 'AGTCGTTAGAT'
    pattern = 'CGTTA'
    rb.find(genome, pattern, kamus)
````

*Catatan: Output di atas hanya salah satu metode dari berbagai cara mendesain fungsi hashingnya. Anda tidak harus mengikuti fungsi hashing seperti di atas.*

Ketentuan kompleksitas waktu: $\mathcal{O}(mn)$, dengan average case $\Theta(n)$.

Ketentuan kompleksitas ruang: $\mathcal{O}(m+n)$.

````python
class RabinKarp:
    def myhash(self, string, kamus):
        val = 0
        for i in range(len(string)):
            val += kamus[string[-i-1]]*(4**i)
        return val

    def find(self, genome, pattern, kamus):
        print('Genome')
        print(genome)
        m = len(pattern)
        findval = self.myhash(pattern, kamus)
        print(f'value("{pattern}") = {findval}')

        iter = 1
        cur = self.myhash(genome[:m], kamus)
        print(f'\nIterasi {iter}. substring: {genome[:m]}, value = {cur}')

        for i in range(1, len(genome) - m+1):
            iter += 1
            cur = (cur - kamus[genome[i-1]]*(4**(m-1))) * 4 + kamus[genome[i+m-1]]
            print(f'\nIterasi {iter}. substring: {genome[iter-1:m+iter-1]}, value = {cur}')

            if cur == findval:
                print('Pattern is found!')


if __name__ == '__main__':
    rb = RabinKarp()
    kamus = {'A': 0,
             'G': 1,
             'T': 2,
             'C': 3}
    genome = 'AGTCGTTAGAT'
    pattern = 'CGTTA'
    rb.find(genome, pattern, kamus)
````

## Variasi Masalah Barisan Koin

Dalam tugas ini, Anda akan menyelesaikan sebuah variasi masalah barisan koin. Tugas lab ini akan menguji keterampilan Anda dalam melakukan formulasi DP dan menerjemahkan formulasi tersebut dalam implementasi algoritma.

**Variasi Masalah Barisan Koin**

Masih dengan tugas yang sama, dalam masalah ini, Anda diminta untuk mengumpulkan sebanyak mungkin nominal koin dalam sebuah barisan $N$ buah koin. Perbedaannya, dalam variasi ini, Anda **diperbolehkan** untuk mengambil **dua** buah koin yang bersebelahan, namun Anda **tidak diperbolehkan** mengambil **tiga** buah koin bersebelahan secara berturut-turut.

**Contoh**. Misalkan:

```
N = 6
v = [1, 6, 8, 2, 3, 5]
```
Maka, nilai terbesar yang dapat diperoleh dari barisan koin di atas adalah sebesar 22 dengan mengambil koin indeks 1, 2, 4, dan 5, yaitu 6 + 8 + 3 + 5 = 22.

### Tugas 5.1.1

Untuk memulai formulasi masalah, pertama mari kita mengekstrak relasi rekurens dalam masalah ini.

Misalkan $0\le k\le N-1$. Misalkan pula $M(k)$ menyatakan total nominal koin terbesar yang dapat diperoleh dari $k$ buah koin pertama dari barisan koin yang tersedia.

Untuk mendapatkan relasi rekurens dari $M(k)$, mari kita memikirkan skenario pilihan aksi yang dapat diambil sebagai berikut. Tinjau masalah ketika terdapat $k\ge 3$ buah koin dalam barisan. Berbeda dengan formulasi pada masalah originalnya, dalam variasi ini, strategi kita adalah dengan meninjau **dua buah koin paling kanan**, yaitu koin ke-$(k-2)$ dan koin ke-$(k-1)$:

1. **Koin ke-$(k-1)$ tidak diambil**. 

  * Dalam kasus ini, koin ke-$(k-2)$ masih belum dapat ditentukan statusnya: dapat diambil atau dapat juga tidak diambil. 
  * Artinya, masih tersisa $(k-1)$ koin lainnya yang masih belum dipertimbangkan. 
  * Total koin terbesar yang dapat diambil dari $(k-1)$ koin lainnya tersebut secara rekursif tidak lain adalah $F(k-1)$.
  * Ini berarti bahwa total koin yang terbesar yang dapat diambil ketika aksi ini diambil adalah sebesar ```M(k-1)```.

2. **Koin ke-$(k-1)$ diambil**, namun **Koin ke-$(k-2)$ tidak diambil**
  * Dalam kasus ini, koin ke-$(k-3)$ masih belum dapat ditentukan statusnya: dapat diambil atau dapat juga tidak diambil. 
  * Artinya, masih tersisa $(k-2)$ koin lainnya yang masih belum dipertimbangkan. 
  * Total koin terbesar yang dapat diambil dari $(k-2)$ koin lainnya tersebut secara rekursif tidak lain adalah $F(k-2)$.
  * Ini berarti bahwa total koin yang terbesar yang dapat diambil ketika aksi ini diambil adalah sebesar ```v[k-1]+M(k-2)```.

3. **Koin ke-$(k-1)$ diambil**, dan **Koin ke-$(k-2)$ juga diambil**
  * Karena persyaratan tidak memperbolehkan tiga koin bersebelahan diambil sekaligus, ini berarti bahwa koin ke-$(k-3)$ sudah pasti tidak boleh diambil. 
  * Dengan demikian, tersisa $(k-3)$ koin lainnya yang masih belum dipertimbangkan. 
  * Total koin terbesar yang dapat diambil dari $(k-3)$ koin lainnya tersebut secara rekursif tidak lain adalah $F(k-3)$.
  * Ini berarti bahwa total koin yang terbesar yang dapat diambil ketika aksi ini diambil adalah sebesar ```v[k-1]+v[k-2]+M(k-3)```.

Dari tiga pilihan di atas, kita akan memilih mana di antaranya yang menghasilkan total nominal koin lebih besar. 

**Deskripsi Tugas**: Dengan mempertimbangkan narasi di atas, tuliskan persamaan rekurens secara umum untuk $M(k)$ dalam variasi masalah ini.

*Petunjuk*: Ingat bahwa Anda masih perlu memikirkan bagaimana penyelesaikan kasus basis untuk relasi rekurens tersebut yang belum dibahas pada narasi di atas.

$$
M(k) =\begin{cases}
0, & k=0,\\
v_0, & k=1,\\
v_0 + v1, & k=2,\\
\max\{M(k-1),v[k-1]+M(k-2),v[k-1]+v[k-2]+M(k-3)\}, & k\ge 3.
\end{cases}
$$

Relasi Rekurens nya adalah sebagai berikut:
<br>
M(k): total nominal koin terbesar yang dapat diperoleh dari k buah koin pertama dari barisan koin yang tersedia adalah 0
<br><br>
Untuk k = 0, artinya tidak ada koin yang dapat diambil. Maka dari itu, M(k) = 0
<br>
Untuk k = 1, artinya tersisa satu koin sehingga koin itulah yang diambil maka dari itu, M(k) = 1
<br>
Untuk k = 2, artinya tersisa dua koin yang bisa diambil sehingga kedua koin itu diambil. Maka dari itu, M(k) = v0 + v1
<br>
Untuk k>=3, artinya kita tersisa lebih atau sama dengan tiga koin sehingga terdapat ketiga pilihan seperti yang sudah dijelaskan diatas. 
<br>
Maka dari itu, M(k) = maksimum dari ketiga pilihan diatas yang ditunjukkan pada relasi rekurens diatas

### Tugas 5.1.2

Implementasikan formulasi yang sudah Anda buat pada Tugas 5.1.1 ke dalam bentuk Dynamic Programming secara **bottom up**.

**Ketentuan efisiensi waktu:** $\mathcal{O}(N)$.

**Ketentuan efisiensi ruang:** $\mathcal{O}(N)$.

````python
def variasi_barisan_koin(v):
  n = len(v)

  # Inisialisasi tabel untuk menyimpan nilai M(k)
  # O(1) Time
  memo = [None] * (n+1)

  # Mengisi nilai tabel memo untuk base case
  # O(1) Time
  memo[0] = 0
  memo[1] = v[0]
  memo[2] = v[0]+v[1]

  # Dynamic Programming Bottom Up O(n) time
  # memo(k-1) => koin ke k-1 tidak diambil
  # v[k-1]+memo[k-2] => koin ke k-1 diambil dan koin ke k-2 tidak diambil
  # v[k-1]+v[k-2]+memo[k-3] => koin ke k-1 dan k-2 diambil
  for k in range(3, n+1):
      memo[k] = max(memo[k-1], v[k-1]+memo[k-2], v[k-1]+v[k-2]+memo[k-3])

  print('Tabel Memoisasi')
  print(memo, '\n')
  return f'Total nominal koin terbanyak yang dapat diambil adalah {memo[n]}'


if __name__ == '__main__':
  v = [1, 6, 8, 2, 3, 5]
  print(variasi_barisan_koin(v))
````

Kompleksitas Waktu: O(n)
<br>
Algortima mengisi tabel memoisasi yang panjangnya n (banyaknya koin) dan setiap kali mengisi tabel memoisasi, algoritma mencatat M(k) terbaik yaitu maksimum dari ketiga pilihan yang memungkinkan dan kompleksitas waktu per mengisi adalah O(1). Sehingga kompleksitas waktunya adalah O(n).
<br><br>
Kompleksitas Ruang: O(n)
<br>
Memori yang dibutuhkan algoritma ini hanya array sepanjang n (banyaknya koin) sehingga kompleksitas memorinya adalah O(n).

````python
def variasi_barisan_koin_dengan_tracing(v):
  n = len(v)

  # Inisialisasi tabel untuk menyimpan nilai M(k) dan trace[k]
  # O(1) Time

  memo = [None] * (n+1)
  trace = [None] * (n+1)

  # Mengisi nilai tabel memo untuk base case
  # O(1) Time
  memo[0] = 0
  memo[1] = v[0]
  memo[2] = v[0]+v[1]

  # Mengisi nilai tabel trace untuk base case
  # Kode 1 = Ambil 1 Koin
  # Kode 2 = Ambil 2 Koin
  # Kode 3 = Ambil 3 Koin
  trace[1] = 1
  trace[2] = 2

  # Mengisi nilai tabel M dan trace untuk kasus dengan sifat rekursif
  for k in range(3, n+1):
      # koin ke (k-1) tidak diambil
      option_1 = memo[k-1]
      # koin ke (k-1) diambil dan koin ke (k-2) tidak diambil
      option_2 = v[k-1] + memo[k-2]
      # koin ke (k-1) dan (k-2) diambil
      option_3 = v[k-1] + v[k-2] + memo[k-3]

      # Option 1 yang terbaik
      if option_1 >= option_2 and option_1 >= option_3:
          memo[k] = option_1
          trace[k] = 3
      # Option 2 yang terbaik
      elif option_2 >= option_1 and option_2 >= option_3:
          memo[k] = option_2
          trace[k] = 1
      # Option 3 yang terbaik
      else:
          memo[k] = option_3
          trace[k] = 2

##############################################
#             ALGORITMA PELACAKAN            #
##############################################
  coins = []
  k = n
  while k > 0:
    # ambil 1 koin
      if trace[k] == 1:
          coins.append(k-1)
          k -= 2    
    # ambil 2 koin
      elif trace[k] == 2:
          coins.append(k-1)
          coins.append(k-2)
          k -= 3
      else:
        # ga ambil apa"
          k -= 1

  print('Tabel Memoisasi ')
  print(memo)
  print('Tabel Trace')
  print(trace, '\n')
  return memo[n], coins


if __name__ == '__main__':
  v = [1, 6, 8, 2, 3, 5]
  max_coin, chosen_coins = variasi_barisan_koin_dengan_tracing(v)
  print(
      f'Total nominal koin terbanyak yang dapat diambil adalah {max_coin}.')
  print('Koin-koin yang diambil adalah sebagai berikut:')
  # Membalik urutan agar koin terbaca dari kiri ke kanan
  chosen_coins.reverse()
  coin_list = []
  for coin_index in chosen_coins:
      coin = v[coin_index]
      coin_list.append(coin)
      print(f'\tKoin ke-{coin_index}, senilai {coin}')
  nilai = sum(coin_list)
  coin_list = map(str, coin_list)
  print(' + '.join(coin_list)+f' = {nilai}')
````

````python
def tantangan_2(H):
  N = len(H)
  M = H[0]
  for i in range(1,N):
    M_ = [None]*N
    for j in range(N):
      a,b,c = 0,0,0
      b = M[j]
      if j != 0: #kalau sudah di ujung kiri tidak boleh ambil diagonal kiri belakang
        a = M[j-1] 
      if j != N-1: #kalau sudah di ujung kanan tidak boleh ambil diagonal kanan belakang
        c = M[j+1]
      M_[j] = max(a,b,c) + H[i][j]
    M = M_
  return max(M)

H = [[7, 2, 3, 1],
     [1, 2, 8, 9],
     [1, 4, 6, 5],
     [9, 1, 3, 2]]

tantangan_2(H)
````

* Total nilai item $x_0v_0 + x_1v_1 +\dots + x_{N-1}v_{N-1} $ dimaksimumkan.

**Ketentuan efisiensi waktu:** $\mathcal{O}(MC)$ di mana $M=\sum\limits_{i}m_i$.

**Ketentuan efisiensi ruang:** $\mathcal{O}(NC)$.

````python
def knapsack_table(n, c_max, v, w, m):
    M = [[0]*(c_max+1)] + [[None]*(c_max+1) for i in range(n)]
    for k in range(1, n+1):
        for c in range(c_max+1):
            max_val = []
            for i in range(m[k-1]+1):
                if c-(i*w[k-1]) >= 0:
                    max_val.append(v[k-1]*i + M[k-1][c-(i*w[k-1])])

            M[k][c] = max(max_val)   

    print('Tabel Memoisasi')
    for i in M:
      print(i)
    hasil = M[n][c_max] 
    return hasil


if __name__ == '__main__':
    v = [75, 30, 600, 1000, 250, 400, 800, 750]
    w = [1, 2, 8, 12, 4, 3, 7, 6]
    m = [2, 1, 4, 1, 4, 2, 3, 3]
    v = [10,5,1]
    w = [7,3,2]
    m = [2,2,1]
    c = 15
    print(f'\nMaka total penghasilan terbesar yang dapat diraih adalah sebesar Rp{knapsack_table(len(v), c, v, w, m)}.')
````

**Ketentuan efisiensi waktu:** $\mathcal{O}(MN)$ di mana $M=|x|$ dan $N=|y|$.

**Ketentuan efisiensi ruang:** $\mathcal{O}(MN)$ di mana $M=|x|$ dan $N=|y|$.

````python
def edit_distance(x, y, thevar):
    # Ukuran Memo 
    memo = [[0]*(len(y)+1) for i in range(len(x)+1)]
    # Isi Baris Pertama Memo
    for i in range(len(memo[0])):
        memo[0][i] = i

    # Isi Kolom Pertama Memo
    for i in range(len(memo)):
        memo[i][0] = i

    for i in range(1, len(memo)):
        for j in range(1, len(memo[0])):
            if x[i-1] == y[j-1]:
                memo[i][j] = memo[i-1][j-1]
            else:
                from_a = thevar.find(x[i-1])
                to_b = thevar.find(y[j-1])
                s1 = replace_cost[from_a][to_b] + memo[i-1][j-1]
                s2 = 3 + memo[i][j-1]
                s3 = 2 + memo[i-1][j]
                memo[i][j] = min(s1, s2, s3)

    print('Tabel Memoisasi')
    for i in memo:
        print(i)
    return f'\nTotal Edit Distance = {memo[-1][-1]}'


if __name__ == '__main__':
    x = "AACAGTTACC"
    y = "TAAGGTCA"
    thevar = "AGTC"
    x = "AAT"
    y = "TA"

    replace_cost = [
        [0, 4, 2, 1],
        [1, 0, 1, 6],
        [2, 1, 0, 1],
        [2, 2, 3, 0]
    ]
    print('\n'+edit_distance(x, y, thevar))
````

**Ketentuan efisiensi ruang:** $\mathcal{O}(N+M)$.


Petunjuk Implementasi: Gunakan *Priority Queue* saat melakukan proses eksplorasi mirip BFS.

````python
from queue import PriorityQueue

##### Graf Soal #####
N = 6
M = edge_list = [
  [0, 1, 8],
  [0, 2, 6],
  [2, 5, 4],
  [2, 3, 2],
  [0, 3, 3],
  [1, 5, 5],
  [3, 4, 1],
  [1, 3, 3],
  [3, 5, 2],
  [2, 4, 2],
  [4, 5, 2]]
  
##### Ubah Edgelist jadi AdjDict #####
adjgraf = {i:[] for i in range(N)}
print(adjgraf)
for u,v,weight in M:
  adjgraf[u].append((weight,v))
  adjgraf[v].append((weight,u))
print(f'adjgraf = {adjgraf}')

##### Siapkan variabel untuk simpan hasil #####
Root_eccentricity = 0
total_weight = 0
MST = []

##### Inisiasi Queuenya, penelusuran pertama itu node R #####
visited = set()
R = 3
q = PriorityQueue()
cur = R

##### Cari semua tetangga dari R dan masukan kedalam Priority Queue ####
for weight,adj in adjgraf[cur]:
    #priority queue akan mengurutkan antrian berdasarkan bobot dari yang paling kecil

  q.put((weight,1,adj,cur)) 

visited|={cur}

##### Lakukan penelusuran BFS menggunakan Priority Queue #####
while not q.empty():
    #ambil data node dari priority queue
  w,dist,cur,before = q.get() 

    #hanya ambil node yang tidak pernah ditelusuri sebelumnya supaya tidak ada cycle
  if cur not in visited: 
      #Root eccentricity itu jarak terjauh dari node ke leaf
    Root_eccentricity = max(dist,Root_eccentricity) 
    #masukan sisi kedalam MST
    MST.append([cur,before]) 
    #update total weight
    total_weight+=w 
    #tandai node yang ditelusuri sebagai sudah dikunjungi
    visited|={cur} 
    #masukan tetangga node tersebut kedalam priority queue
    for weight,adj in adjgraf[cur]: 
      q.put((weight,Root_eccentricity+1,adj,cur))

MST.sort()

print("edge list:")
for i in MST:
  print(i)
print(f"W = {total_weight}")
print(f"Root Eccentricity = {Root_eccentricity}")
````

````python
from queue import PriorityQueue
import math

def ASCII_Encoding(text):
  l,m=[],""
  for i in text:
    l.append(ord(i)) #ubah setiap karakter text menjadi code ASCII
  print(l)

  for i in l:
    m+=bin(i)[2:].rjust(8,"0") #buat code ASCII jadi binary, pastikan ada 8 digit, lalu disatukan 

  print("1. ASCII Encoding:")
  print(m)
  print("length :", len(m))
  return m,len(m)

def Frequency_Analysis(text):
  print("2. Frequency Analysis:")
  len_text = len(text)
  freq = {i : 0 for i in set(text)}
  print(f'Dictionary freq awal = {freq}')
  for i in text: #hitung kemunculan setiap huruf
    freq[i] += 1
  for i in "Lae,dLk niylhgmtrcuoT;Is.fpK": #iterasi setiap karakter unik, urutannya ikut output
    freq[i] = freq[i]/len_text #frekuensi dihitung dengan cara banyak kemungkinan karakter dibagi total 
    print(i,"    :",freq[i])
  print(f'Dictionary freq awal = {freq}')
  return freq


def huffman_code(freq):
  q = PriorityQueue() 

  for i in freq:
    q.put((freq[i],i)) #masukan nilai frekuensi dan karakter ke priorityqueue yang diurutkan dari frekuensi terkecil
  
  huffman_tree = {} #untuk menyimpan huffman tree

  i = 0
  while q.qsize() != 1:
    freq_l,left = q.get() #ambil node dengan frekuensi terkecil
    freq_r,right = q.get() #ambil node dengan frekuensi terkecil kedua

    i+=1
    key = f"v{i}"
    if q.empty():
      key = "root"
    huffman_tree[key] = [("0",left),("1",right)]
    q.put((freq_l+freq_r,key))
  huffman_dict = {}

  queue = [("root","")]
  while queue:
    cur,bin_huff = queue.pop(0)
    
    for bin, node in huffman_tree[cur]:
      if len(node) == 1:
        huffman_dict[node] = bin_huff+bin
      else:
        queue.append((node,bin_huff+bin))

  print("3. Huffman Coding Dictionary:")
  for i in "Lae,dLk niylhgmtrcuoT;Is.fpK":
    print(i,"   :",huffman_dict[i])
  
  return huffman_dict,huffman_tree

def huffman_decoding(text,huffman_dict):
  ans = ""
  for i in text:
    ans+=huffman_dict[i]
  
  print("4. Huffman Encoding:")
  print(ans)
  print("length:",len(ans))
  return ans
  
def compress_ratio(freq,huffman_dict):
  fixed_length = math.ceil(math.log2(len(huffman_dict)))
  rata_rata_panjang_code_word = 0
  for i in huffman_dict:
    rata_rata_panjang_code_word+= len(huffman_dict[i])*freq[i]
  compress_rate = round(((fixed_length-rata_rata_panjang_code_word)/fixed_length)*100,2)

  print("5. Compress Ratio:")
  print(f"fixed-length codeword number of bits: {fixed_length}")
  print(f"huffman coding average codeword number of bits: {rata_rata_panjang_code_word}")
  print("")
  print(f"Compress ratio : {compress_rate}%")



text = "Lead, kindly Light, amid the encircling gloom, Lead thou me on; The night is dark, and I am far from home, Lead thou me on. Keep thou my feet; I do not ask to see The distant scene; one step enough for me."
ASCII_Encoding(text)
print("\n")
freq = Frequency_Analysis(text)
print("\n")
huffman_dict,huffman_tree = huffman_code(freq)
print("\n")
decoding = huffman_decoding(text,huffman_dict)
print("\n")
compress_ratio(freq,huffman_dict)
````
